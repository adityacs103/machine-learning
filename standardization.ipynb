```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('https://raw.githubusercontent.com/campusx-official/100-days-of-machine-learning/main/day24-standardization/Social_Network_Ads.csv')
```

```python
df
```

```python
df.drop(columns='User ID',inplace=True)
```

```python
df.drop(columns='Gender',inplace=True)
```

```python
df
```

```python
x=df[['Age','EstimatedSalary']]
```

```python
y=df['Purchased']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=10)
```

```python
x_train.head()
```

```python
x_train.shape,x_test.shape
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scalar=StandardScaler()
```

```python
scalar.fit(x_train)
```

```python
x_train_scaled=scalar.transform(x_train)
x_test_scaled=scalar.transform(x_test)
```

```python
x_train_scaled
```

```python
x_train_scaled=pd.DataFrame(x_train_scaled,columns=x_train.columns)
```

```python
x_train_scaled
```

```python
x_test_scaled=pd.DataFrame(x_test_scaled,columns=x_test.columns)
```

```python
x_test_scaled
```

```python
np.round(x_train_scaled.describe(),1)
```

```python
plt.scatter(x_train['Age'],x_train['EstimatedSalary'])
plt.title('beore scaling')
```

```python
plt.scatter(x_train_scaled['Age'],x_train_scaled['EstimatedSalary'],color='red')
plt.title("After scaling")
```

```python
sns.kdeplot(x_train)
```

```python
sns.kdeplot(x_train_scaled)
```

```python
sns.distplot(x_train)
```

```python
sns.distplot(x_train_scaled)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr=LogisticRegression()
lr_scaled=LogisticRegression()
```

```python
lr.fit(x_train,y_train)
lr_scaled.fit(x_train_scaled,y_train)
```

```python
y_pred=lr.predict(x_test)
y_pred_scaled=lr_scaled.predict(x_test_scaled)
```

```python
from sklearn.metrics import accuracy_score
```

```python
print("Actual",accuracy_score(y_test,y_pred)*100)
print("scaled",accuracy_score(y_test,y_pred_scaled)*100)
```
