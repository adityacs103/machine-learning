```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
```

```python
num_columns=list(X.select_dtypes(include=['int','float']).columns)
```

```python
cat_columns=list(X.select_dtypes(include='category').columns)
```

```python
num_columns
```

```python
cat_columns
```

```python
# num_pipe=make_pipeline(SimpleImputer(strategy='median'),
#                         StandardScaler())
num_pipe=Pipeline(steps=[

    ("imputer",SimpleImputer(strategy='median')),

    ('scaling',StandardScaler())
    
])
```

```python
num_pipe
```

```python
# cat_pipe=make_pipeline(SimpleImputer(strategy='most_frequent'),
#                         OneHotEncoder())
cat_pipe=Pipeline(steps=[
    ("imputer",SimpleImputer(strategy='most_frequent')),
    ('onehot',OneHotEncoder())

])
```

```python
cat_pipe
```

```python
transformer=ColumnTransformer([
    ("num_pipe",num_pipe,num_columns),
    ("cat_pipe",cat_pipe,cat_columns) 
])
```

```python
transformer
```

```python
x_train
```

```python
y_train
```

```python
x_train=transformer.fit_transform(x_train)
```

```python
x_test=transformer.fit_transform(x_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
```

```python
models={
    "Random_forest":RandomForestClassifier(),
    "Logistic_regression":LogisticRegression(),
    "Decision_tree":DecisionTreeClassifier()
}
```

```python
def evaluate_model(x_train,x_test,y_train,y_test,models):
    for model_name,model in models.items():
        model.fit(x_train,y_train)
        y_pred=model.predict(x_test)
        print(f"{model_name}: {round(accuracy_score(y_test,y_pred)*100)}")
```

```python
evaluate_model(x_train,x_test,y_train,y_test,models)
```

```python
param={
    "n_estimators":[10,15,20,25],
    'criterion':["gini","entropy"],
    "max_depth":[3,5,7]
}
```
